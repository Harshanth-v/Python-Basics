```python
def exponent(base, exp):
    result = 1
    for i in range(exp):
        result *= base
    return result

print("Result:", exponent(2, 5))
```

```python
primes = []
for num in range(2, 21):
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            break
    else:
        primes.append(num)

print("All primes:", primes)
print("Alternate primes:", primes[::2])
```
