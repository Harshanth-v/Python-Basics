```python
#add two no
def add(a,b):
    return a+b
```

```python
#square of a number
def square(n):
    return n**2
```

```python
#if a no is even or odd
def even_odd(n):
    return "even" if n%2==0 else "odd"
```

```python
#max of 2 no
def max(a,b):
    return a if a>b else b
```

```python
#factorial of a no
def factorial(n):
    fact =1
    for i in range(1,3+1):
        fact*=1
    print(fact)
    
```

```python
#length of a string
def string_length(s):
    return len(s)   
```

```python
#count vowels in a string
def count_vowels(s):
    count=0
    for char in s:
        if char.lower() in 'aeiou':
            count+=1
    return count
```

```python
#reverse a string
def reverse_string(s):
    return s[::-1]
```

```python
#sum of all elements in a list
def sum_list(lst):
    return sum(lst)
```

```python
#largest element in a list
def largest_element(lst):
    return max(lst)
```

```python
#smallest element in a list
def smallest_element(lst):
    return min(lst)
```

```python
#check if a no is prime
def is_prime(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
```

```python
#generate fibonacci series
def fibonacci_series(n):
    fib_series = [0,1]
    for i in range(2,n):
        next_fib = fib_series[i-1] + fib_series[i-2]
        fib_series.append(next_fib)
    return fib_series[:n]   
```

```python
#check if string is palindrome
def is_palindrome(s):
    return s==s[::-1]
```

```python
#celsius to fahrenheit
def celsius_to_fahrenheit(c):
    return (c*9/5)+32
```

```python
#fahrenheit to celsius
def fahrenheit_to_celsius(f):
    return (f-32)*5/9
```

```python
#absolute value of a no
def absolute_value(n):
    return abs(n)
```

```python
#sum of digits in a no
def sum_of_digits(n):
    return sum(int(digit) for digit in str(n))
```

```python
#product of digits in a no
def product_of_digits(n):
    product=1
    for digit in str(n):
        product*=int(digit)
    return product
```

```python
#gcd of 2 no
def gcd(a,b):
    while b:
        a,b=b,a%b
    return a
```

```python
#lcm of 2 no
def lcm(a,b):
    return abs(a*b)//gcd(a,b)
```

```python
#check is a number is Armstrong or not
def is_armstrong(n):
    num_str = str(n)
    num_digits = len(num_str)
    sum_of_powers = sum(int(digit)**num_digits for digit in num_str)
    return sum_of_powers == n
```

```python
#second largest element in a list
def second_largest(lst):
    unique_lst = list(set(lst))
    unique_lst.sort()
    if len(unique_lst)<2:
        return None
    return unique_lst[-2]
```

```python
#count the number of words in a string
def count_words(s):
    words = s.split()
    return len(words)
```

```python
#return unique elements in a list
def unique_elements(lst):
    return list(set(lst))
```

```python
#the power of a no
def power(x,y):
    return x**y
```

```python
#print the multiplication table of a no
def multiplication_table(n):
    for i in range(1,11):
        print(f"{n} x {i} = {n*i}")
```

```python
#count uppercase and lowercase letters in a string
def count_upper_lower(s):
    upper_count=0
    lower_count=0
    for char in s:
        if char.isupper():
            upper_count+=1
        elif char.islower():
            lower_count+=1
    return upper_count, lower_count
```

```python
#find the average of a list of numbers
def average(lst):
    return sum(lst)/len(lst)
```

```python
#factorial using recursion
def factorial_recursive(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial_recursive(n-1)
```
